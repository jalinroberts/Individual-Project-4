#### 4.	Import car-sales.csv  into a DataFrame called car_sales and view.

```python
# Load the CSV file into a DataFrame
car_sales = pd.read_csv("car-sales.csv")

# View the DataFrame
print(car_sales)
```

#### 4a.) Try to plot the 'Price' column using the plot() function

#### The reason this cannot be plotted is because the 'Price' column contains symbols, so it needs to be converted to a float type.

```python
# Remove the dollar sign and comma from the 'Price' column and convert it to a float type
car_sales['Price'] = car_sales['Price'].str.replace('$','', regex=False).str.replace(',','', regex=False).astype(float)

# Plot the 'Price' column
car_sales['Price'].plot()

# Add a label for the x axis
plt.xlabel("Index")

# Add a label for the y axis
plt.ylabel("Price")

# Show the plot
plt.show()
```

##### 4b.) Try to plot the"Odometer (KM)" column using the plot() function

```python
# Plot the 'Odometer (KM)' column
car_sales['Odometer (KM)'].plot()

# Add a label for the x axis
plt.xlabel("Index")

# Add a label for the y axis
plt.ylabel("Odometer (KM)")

# Show the plot
plt.show()
```

#### 5.) Remove all the symbols from the 'Price' column and convert and update it to 64-bit int numbers using astype('int64')

```python
# Convert the 'Price' column to strings
car_sales['Price'] = car_sales['Price'].astype(str)

# Remove the '$' symbol from the 'Price' column
car_sales['Price'] = car_sales['Price'].str.replace('$', '', regex=False)

# Convert the 'Price' column to a float
car_sales['Price'] = car_sales['Price'].astype(float)


# Convert the 'Price' column to 64-bit int numbers
car_sales['Price'] = car_sales['Price'].astype('int64')

print(car_sales)
```

#### 6.) Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column. Look up the cumsum() function.

```python
# Calculate the cumulative sum of the 'Price' column and store it in a new column called 'Total Sales'
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Print the updated dataframe
print(car_sales)
```

#### 7.) Add a column called 'Sale Date' which lists a series of successive dates starting from today. Check out "date_range" function and keep the periods=len(car_sales)

```python
# Create a series of dates starting from today
today = pd.Timestamp('today')
sale_dates = pd.date_range(today, periods=len(car_sales), freq='D')

# Add the series of dates to the 'Sale Date' column of the dataframe
car_sales['Sale Date'] = sale_dates

# Print the updated dataframe
print(car_sales)
```

#### 8.) Use matplotlib to plot the 'Price' column versus the 'Total Sales' column

```python
# Plot the 'Price' column versus the 'Total Sales' column
plt.plot(car_sales['Price'], car_sales['Total Sales'])

# Add labels and title to the plot
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs. Total Sales')

# Show the plot
plt.show()
```

#### 9.) Create a matplotlib scatter plot of the 'Odometer (KM)' vs 'Price column.

```python
# Plot the 'Odometer (KM)' versus 'Price' column as a scatter plot
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])

# Add labels and title to the plot
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer (KM) vs Price')

# Show the plot
plt.show()
```

#### 10.) Create a matplotlib bar graph of the 'Make' vs 'Odometer (KM)' columns

```python
# Plot the 'Make' versus 'Odometer (KM)' column as a bar graph
plt.bar(car_sales['Make'], car_sales['Odometer (KM)'])

# Add labels and title to the plot
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.title('Make vs Odometer (KM)')

# Show the plot
plt.show()
```

#### 11.) Create a histogram of the 'Make' Column

#### 12.) Create a subplot of multiple plots (step 8, 9, 10, and 11)

```python
# Create subplots
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

# Plot the 'Price' column versus the 'Total Sales' column
ax[0][0].plot(car_sales['Price'], car_sales['Total Sales'])

# Add labels and title to the plot
ax[0][0].set_xlabel('Price')
ax[0][0].set_ylabel('Total Sales')
ax[0][0].set_title('Price vs. Total Sales')

# Plot the 'Odometer (KM)' versus 'Price' column as a scatter plot
ax[0][1].scatter(car_sales['Odometer (KM)'], car_sales['Price'])

# Add labels and title to the plot
ax[0][1].set_xlabel('Odometer (KM)')
ax[0][1].set_ylabel('Price')
ax[0][1].set_title('Odometer (KM) vs Price')

# Plot the 'Make' versus 'Odometer (KM)' column as a bar graph
ax[1][0].bar(car_sales['Make'], car_sales['Odometer (KM)'])

# Add labels and title to the plot
ax[1][0].set_xlabel('Make')
ax[1][0].set_ylabel('Odometer (KM)')
ax[1][0].set_title('Make vs Odometer (KM)')

# Plot a histogram of the 'Make' column
ax[1][1].hist(car_sales['Make'])

# Add labels and title to the plot
ax[1][1].set_xlabel('Make')
ax[1][1].set_ylabel('Frequency')
ax[1][1].set_title('Histogram of Make')

# Show the subplots
plt.show()
```
